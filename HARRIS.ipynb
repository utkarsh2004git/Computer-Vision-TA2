### 4. Implement the Harris corner detector to find and visualize corners in a grayscale image.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('corner.jpg')
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
```

```python
dst = cv2.cornerHarris(gray_img,2, 5, 0.04)
dst = cv2.dilate(dst, None,iterations=3)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
thresh = 0.01 * dst.max()
corner_img = np.copy(img)
for i in range(0, dst.shape[0]):
    for j in range(0, dst.shape[1]):
        if dst[i, j] > thresh:
            cv2.circle(corner_img, (j, i), 1, (255, 0, 0), 1)
plt.imshow(corner_img)
```
